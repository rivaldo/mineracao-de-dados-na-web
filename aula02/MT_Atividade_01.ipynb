```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
```

```python
lista_temas = ["covid+saude"]
```

```python
print(f'https://repositorio.ufpe.br/simple-search?location=123456789%2F50&query={lista_temas[0]}')
```

```python
response = requests.get(f'https://repositorio.ufpe.br/simple-search?location=123456789%2F50&query={lista_temas[0]}', allow_redirects=True)
```

```python
with open('output_query_request.html', 'wb') as fb:
  fb.write(response.content)
print(response.text)
```

```python
from bs4 import BeautifulSoup

soup = BeautifulSoup(response.text.strip(), 'lxml')
tags = soup.find_all(name='table')
tag_table = tags[0]
print(tag_table)
```

```python
tr = soup.find_all(name='tr')
tr.pop(0)
print(tag_table)
```

```python
href_link = []
for x in tr:
  link = x.find_all('td')[1].find('a')['href']
  temp = f'https://repositorio.ufpe.br{link}'
  href_link.append(temp)
  print(link)
```

```python
href_link
```

```python
itens = []
for url in href_link:
  print(url)
  response = requests.get(url, allow_redirects=True)
  soup = BeautifulSoup(response.content, 'lxml')
  titulo = soup.find(name='meta', attrs={'name': 'DC.title'}).attrs['content']
  criador = soup.find(name='meta', attrs={'name': 'DC.creator'}).attrs['content']
  abstract = soup.find(name='meta', attrs={'name': 'DCTERMS.abstract'}).attrs['content']
  item = {'titulo': titulo, 'criador': criador, 'abstract': abstract, 'url': url}
  itens.append(item)
```

```python
print(itens)
```

```python
df = pd.DataFrame(itens)
```
