# Baixar páginas HMTL

Utilizei a lib nativa requests 

```python
import requests

#https://repositorio.ufpe.br/simple-search?location=%2F&query=previdencia&rpp=10&sort_by=score&order=desc

response = requests.get('https://repositorio.ufpe.br/simple-search?query=previdencia', allow_redirects=True)
with open('output_query_request.html', 'wb') as fb:
  fb.write(response.content)

print(response.text)
```

Achando os links das Teses

```python
from bs4 import BeautifulSoup

soup = BeautifulSoup(response.text.strip(), 'html.parser')

tags = soup.find_all(name='table')
tag_table = tags[0]
print(tag_table)
```

```python
tags_a = tag_table.find_all(name='a')
link = tags_a[0]['href']
link
```

### Utilizando a lib BeautifulSoup para processamento e manipulação do conteúdo HTML

```python
from bs4 import BeautifulSoup

soup = BeautifulSoup(response.text.strip(), 'html.parser')

print(soup.prettify())
```

```python
soup.find_all(name='meta', attrs={'name': 'DC.subject'})
```

#### Encontrando elementos de interesse para extração de texto

```python
tags = soup.find_all(name='table')
tag_table = tags[0]
print(tag_table)
```

#### Encontrando elementos de interesse para extração do texto

```python
tags_table = soup.find_all(name='table', attrs={'class': 'itemDisplayTable'})
info_table = tags_table[0]
print(info_table)
```

Estruturando informações de valor em um dicionário:

```python
print(info_table)
```

```python
all_titles = info_table.find_all(name='td', attrs={'class': 'dc_title'})
title = all_titles[-1].contents[0]

all_abst = info_table.find_all(name='td', attrs={'class': 'dc_description_abstract'})
abst = all_abst[-1].contents[0]

all_subj = info_table.find_all(name='td', attrs={'class': 'dc_subject'})
subj = all_subj[-1].contents[0]

dict_thesis_info = {}
dict_thesis_info[0] = {
    'title': title,
    'abstract': abst,
    'keywords': subj.split(';'),
    'query': 'previdencia'
}
for k, v in dict_thesis_info[0].items():
  print(k)
  print(v)
  print()
# print(dict_thesis_info)
```

#### Encontrando link com arquivo PDF da Tese

```python
tags_table_link = soup.find_all(name='table', attrs={'class': 'panel-body'})[0]

print(tags_table_link.td.a['href'])

pdf_file_link = tags_table_link.td.a['href']

root_url = 'https://repositorio.ufpe.br'
url_pdf_link = root_url + pdf_file_link

print(url_pdf_link)
```

#### Baixando arquivo PDF da tese

```python
response = requests.get(url_pdf_link, allow_redirects=True)
with open('thesis_1.pdf', 'wb') as fb:
    fb.write(response.content)
```

# Prática com site do Banco Central

https://www.bcb.gov.br/api/servico/sitebcb/atascopom/ultimas?quantidade=1000&filtro=

```python
response = requests.get('https://www.bcb.gov.br/api/servico/sitebcb/atascopom/ultimas?quantidade=1000&filtro=', allow_redirects=True)
```

```python
response_list_dict = response.json()['conteudo']
first_link = response_list_dict[0]['LinkPagina']
first_link
```

```python
report_date = first_link.split('/')[-1]
report_date
```

Link para baixar a Ata a partir da Data de Publicação:

```python
from bs4 import BeautifulSoup

# /api/servico/sitebcb/atascopom/principal?filtro=IdentificadorUrl%20eq%20%27<data>%27
pattern_link = "/api/servico/sitebcb/atascopom/principal?filtro=IdentificadorUrl%20eq%20%27<data>%27"
response_first_link = requests.get('https://www.bcb.gov.br'+pattern_link.replace('<data>', report_date), allow_redirects=True)
# print(response_first_link.json()['conteudo'][0])

string_div_atacompleta = response_first_link.json()['conteudo'][0]['OutrasInformacoes']
soup = BeautifulSoup(string_div_atacompleta, 'html.parser')

soup.prettify()
```
